```python
def biological_neuron(signals,threshold=1.0):
    soma=sum(signals)
    if soma>=threshold:
        return 1
    return 0
sig=[0.3,0.4,0.5]
print("Neuron Op:",biological_neuron(sig))
```

```python
import numpy as np
def artificial_neuron(inp,w,bias=0):
    y_sum=np.dot(inp,w)+bias
    return y_sum
x=np.array([1,0,1])
w=np.array([0.5,-0.6,0.2])
print("Summation Output:",artificial_neuron(x,w,bias=0.1))
```

```python
import math
def identity(x):
    return x
def step(x):
    return 1 if x>=0 else 0
def threshold(x,theta=0.5):
    return 1 if x>=theta else 0
def relu(x):
    return max(0,x)
def sigmoid(x,k=1):
    return 1/(1+math.exp(-k*x))
def bipolar_sigmoid(x):
    return (2/(1+math.exp(-x)))-1
def tanh_n(x):
    return math.tanh(x)
val=[-2,-11,0,0.5,1,2]
print(val)

print("Identity function")
for v in val:
    print(v,"->",identity(v))
print("Step function")
for v in val:
    print(v,"->",step(v))
print("Threshhold function")
for v in val:
    print(v,"->",threshold(v))
print("Relu function")
for v in val:
    print(v,"->",relu(v))
print("Sigmoid function")
for v in val:
      print(v,"->",sigmoid(v))
print("Bipolar Sigmoid function")
for v in val:
    print(v,"->",bipolar_sigmoid(v))
print("tanh function")
for v in val:
    print(v,"->",tanh_n(v))
```

```python
#Rosenblatt
import numpy as np
class Perceptron:
    def __init__(self,lr=0.1,epochs=10):
        self.lr=lr
        self.epochs=epochs
    def activation(self,x):
        return 1 if x>=0 else -1
    def fit(self,x,y):
        self.weights=np.zeros(x.shape[1])
        self.bias=0
        for _ in range(self.epochs):
            for i in range(len(x)):
                linear=np.dot(x[i],self.weights)+self.bias
                y_pred=self.activation(linear)
                update=self.lr*(y[i]-y_pred)
                self.weights+=update*x[i]
                self.bias+=update
    def predict(self,x):
        linear=np.dot(x,self.weights)+self.bias
        return self.activation(linear)
x=np.array([[2,3],[1,1],[2,1],[-1,-2]])
y=np.array([1,2,-1,-1])
p=Perceptron()
p.fit(x,y)
print("Prediction",p.predict(np.array([2,2])))
```

```python
#Rosenblatt
import numpy as np
class Adaline:
    def __init__(self,lr=0.1,epochs=10):
        self.lr=lr
        self.epochs=epochs
    def activation(self,x):
        return 1 if x>=0 else -1
    def fit(self,x,y):
        self.weights=np.zeros(x.shape[1])
        self.bias=0
        for _ in range(self.epochs):
            for i in range(len(x)):
    
                y_sum=np.dot(x[i],self.weights)
                error=y[i]-y_sum
                self.weights+=self.lr*error*x[i]
                self.bias+=self.lr*error
    def predict(self,x):
        y_sum=np.dot(x,self.weights)+self.bias
        return 1 if y_sum>=0 else-1
x=np.array([[2,3],[1,1],[2,1],[-1,-2]])
y=np.array([1,2,-1,-1])
a=Adaline()
a.fit(x,y)
print("Prediction",a.predict(np.array([2,2])))
```
